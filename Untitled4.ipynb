```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, _), (_, _) = tf.keras.datasets.mnist.load_data()

# normalize images
x_train = x_train / 255.0
x_train = x_train.reshape(-1, 28, 28, 1).astype("float32")

BATCH_SIZE = 256
dataset = tf.data.Dataset.from_tensor_slices(x_train).shuffle(60000).batch(BATCH_SIZE)
```

```python
generator = tf.keras.Sequential([
    layers.Dense(128, activation="relu", input_shape=(100,)),
    layers.Dense(28*28, activation="sigmoid"),
    layers.Reshape((28,28,1))
])
```

```python
discriminator = tf.keras.Sequential([
    layers.Flatten(input_shape=(28,28,1)),
    layers.Dense(128, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

discriminator.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
discriminator.trainable = False

gan = tf.keras.Sequential([generator, discriminator])
gan.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
EPOCHS = 10
noise_dim = 100

for epoch in range(EPOCHS):
    for real_images in dataset:

        # train discriminator
        noise = np.random.normal(0,1,(real_images.shape[0], noise_dim))
        fake_images = generator.predict(noise, verbose=0)

        real_labels = np.ones((real_images.shape[0],1))
        fake_labels = np.zeros((real_images.shape[0],1))

        discriminator.trainable = True
        discriminator.train_on_batch(real_images, real_labels)
        discriminator.train_on_batch(fake_images, fake_labels)

        # train generator
        noise = np.random.normal(0,1,(real_images.shape[0], noise_dim))
        misleading_labels = np.ones((real_images.shape[0],1))

        discriminator.trainable = False
        gan.train_on_batch(noise, misleading_labels)

    print("Epoch", epoch+1, "done")
```

```python
noise = np.random.normal(0,1,(16,100))
generated = generator.predict(noise)

plt.figure(figsize=(4,4))
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.imshow(generated[i].reshape(28,28), cmap="gray")
    plt.axis("off")
plt.show()
```
